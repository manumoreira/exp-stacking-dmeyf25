# Notebook básica

```r
# ============================================================================
# PARÁMETROS - Esta celda DEBE tener tag "parameters"
# ============================================================================

numero <- 10
nombre_ventana <- "test"
```

```r
cat("=== Notebook dummy ejecutándose ===\n")
cat("Parámetro recibido - numero:", numero, "\n")
cat("Parámetro recibido - nombre_ventana:", nombre_ventana, "\n\n")

# Cálculo simple: sumar 100 al número
resultado <- numero + 100

cat("Calculo: ", numero, " + 100 = ", resultado, "\n")
```

```r
# Crear directorio si no existe
if (!dir.exists("./resultados")) {
  dir.create("./resultados")
  cat("✓ Directorio ./resultados creado\n")
}

# Guardar resultado en archivo
nombre_archivo <- paste0("./resultados/resultado_", nombre_ventana, ".txt")

writeLines(
  c(
    paste("Ventana:", nombre_ventana),
    paste("Numero original:", numero),
    paste("Resultado:", resultado),
    paste("Timestamp:", Sys.time())
  ),
  nombre_archivo
)

cat("✓ Resultado guardado en:", nombre_archivo, "\n")
```

```r
cat("\n=== Ejecución completada exitosamente ===\n")
cat("Ventana:", nombre_ventana, "- DONE\n")
```
