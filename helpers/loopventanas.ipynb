```r
library(lubridate)

# Fechas inicial y final
current_future <- as.Date("2019-08-01")
end_future <- as.Date("2021-06-01")

while(current_future <= end_future) {
  # Formatear como YYYYMM
  future_str <- format(current_future, "%Y%m")
  
  # Generar meses de entrenamiento (ventana de 6 meses)
  training_months <- character(6)
  for(i in 1:6) {
    training_date <- current_future %m-% months(i) - 1
    training_months[7-i] <- format(training_date, "%Y%m")
  }
  
  cat("PARAM$prod$future <- c(", future_str, ")\n\n")
  cat("PARAM$prod$training <- c(\n")
  cat("  ", paste(training_months, collapse = ", "), "\n")
  cat(")\n\n")
  cat("PARAM$prod$undersampling <- 0.05\n\n")
  cat("# --- Next version ---\n\n")
  
  # Mover al siguiente mes
  current_future <- current_future %m+% months(1)
}
```
