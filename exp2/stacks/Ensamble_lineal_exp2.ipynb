```r
library(data.table)

# --- 1. Configuración de Archivos ---
archivo_rank <- "predicciones rank.csv"
archivo_feat <- "predicciones_feat.csv"
archivo_ipc  <- "predicciones_ipc.txt"

# Cantidad de envíos para el corte
CANTIDAD_ENVIOS <- 11000

dt_rank <- fread(archivo_rank, select = c("numero_de_cliente", "prob"))
setnames(dt_rank, "prob", "prob_rank")

dt_feat <- fread(archivo_feat, select = c("numero_de_cliente", "prob"))
setnames(dt_feat, "prob", "prob_feat")

dt_ipc  <- fread(archivo_ipc, select = c("numero_de_cliente", "prob"))
setnames(dt_ipc, "prob", "prob_ipc")

dt_ensamble <- merge(dt_rank, dt_feat, by = "numero_de_cliente")
dt_ensamble <- merge(dt_ensamble, dt_ipc, by = "numero_de_cliente")

dt_ensamble[, prob_promedio := (prob_rank + prob_feat + prob_ipc) / 3]

setorder(dt_ensamble, -prob_promedio)

dt_ensamble[, Predicted := 0L]
dt_ensamble[1:CANTIDAD_ENVIOS, Predicted := 1L]


nombre_salida <- "kaggle_ensamble_promedio_simple.csv"

fwrite(dt_ensamble[, .(numero_de_cliente, Predicted)],
       file = nombre_salida,
       sep = ",")
```
